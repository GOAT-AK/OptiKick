### Importing & Loading what we need

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

# Loading our data
df = pd.read_csv("Graduation/Expanded_Biometric_Static.csv")
```

```python
def visualize_and_check_outliers(df, numeric_columns):
    """
    This function plots histograms, boxplots, and calculates summary statistics
    for numeric columns to identify outliers and check distribution.
    """
    for col in numeric_columns:
        print(f"Feature: {col}")
        print("Summary Statistics:")
        print(df[col].describe())
        print("\n")

        # Plot histogram to visualize the distribution
        plt.figure(figsize=(12, 6))
        sns.histplot(df[col], kde=True, bins=20, color='blue')
        plt.title(f"Distribution of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()

        # Plot boxplot to check for outliers
        plt.figure(figsize=(12, 6))
        sns.boxplot(x=df[col], color='orange')
        plt.title(f"Boxplot of {col}")
        plt.xlabel(col)
        plt.show()

        # Detect outliers using the IQR method
        Q1 = df[col].quantile(0.25)  # First quartile (25th percentile)
        Q3 = df[col].quantile(0.75)  # Third quartile (75th percentile)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR  # Lower bound
        upper_bound = Q3 + 1.5 * IQR  # Upper bound

        # Identify outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        print(f"Number of Outliers in {col}: {len(outliers)}")
        print("Outliers (if any):")
        print(outliers[[col]])
        print("\n" + "-" * 50 + "\n")

# Identify numeric columns in the dataset
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Call the function to visualize and check outliers
visualize_and_check_outliers(df, numeric_columns)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

# Load data
data = pd.read_csv('Graduation/Expanded_Biometric_Static.csv')

# Drop non-numeric columns (if needed)
numeric_cols = data.select_dtypes(include=[np.number]).columns

# Function to detect and remove outliers using IQR
def remove_outliers_iqr(df, columns):
    """Removes outliers based on IQR for the given columns."""
    initial_size = df.shape[0]  # Store the initial number of rows
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    removed = initial_size - df.shape[0]
    return df, removed

# Function to detect and remove outliers using Z-score
def remove_outliers_zscore(df, columns, threshold=3):
    """Removes outliers based on Z-score for the given columns."""
    initial_size = df.shape[0]
    z_scores = df[columns].apply(zscore)  # Compute Z-scores
    df = df[(z_scores.abs() < threshold).all(axis=1)]
    
    removed = initial_size - df.shape[0]
    return df, removed

# Apply IQR method first
data_iqr, removed_iqr = remove_outliers_iqr(data, ['Fatigue Score', 'Injury Risk', 'Readiness Score'])

# Apply Z-score method after IQR
data_cleaned, removed_zscore = remove_outliers_zscore(data_iqr, ['Fatigue Score', 'Injury Risk', 'Readiness Score'])

# Display results
print(f"Total rows removed by IQR: {removed_iqr}")
print(f"Total rows removed by Z-score: {removed_zscore}")
print(f"Final dataset size: {data_cleaned.shape}")
```

### Dealing With Outliers

* Since the outliers represent less than 5% of our data so we can just remove it safely 

```python
# Select only numerical columns
numeric_cols = data_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Dictionary to store outlier counts
outlier_counts = {}

# Detect and remove outliers using IQR
for col in numeric_cols:
    Q1 = data_cleaned[col].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data_cleaned[col].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range

    lower_bound = Q1 - 1.5 * IQR  # Lower limit
    upper_bound = Q3 + 1.5 * IQR  # Upper limit

    # Find outliers
    outliers = data_cleaned[(data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound)]
    outlier_counts[col] = len(outliers)  # Store the count of outliers

    # Remove outliers
    data_cleaned = data_cleaned[(data_cleaned[col] >= lower_bound) & (data_cleaned[col] <= upper_bound)]

# Print number of outliers in each column
print("Number of outliers removed per column:")
for col, count in outlier_counts.items():
    print(f"{col}: {count}")

# Save the updated DataFrame to a new CSV file
data_cleaned.to_csv('EExpanded_Biometric_Static_New.csv', index=False)

print(" New dataset saved as 'updated_dataset.csv'")
```
